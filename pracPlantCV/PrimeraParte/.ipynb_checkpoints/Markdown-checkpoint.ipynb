# MARKDOWN


Markdown es una herramienta de conversión de texto simple a  HTML. Se usa para publicar en internet en donde HTML se usa frecuentemente: WordPress, Squarespace, Jekyl...  

Se usa en:

- Blogs o páginas web

- Offline para tomar apuntes

- Para escribir libros y luego exportarlo a ePub.


Tiene una sintaxis simple y se puede usar para escribir y dar formato rápidamente sobre todo desde dispositivos móbiles. El formato es compatible con cualquier plataforma que se use.

# ELEMENTOS DE BLOQUE


## PÁRRAFOS Y SALTOS DE LÍNEA

- Los párrafos nuevos se generan mediante una línea en blanco pulsando dos veces la tecla "intro"

- No soporta dobles líneas en blanco y se convierten en una sola línea.

- Para realizar un salto de línea y empezar una línea dentro del mismo párrafo, se pulsa dos veces la barra espaciadora antes de pulsar una vez la tecla "intro".


## ENCABEZADOS

Las almohadillas (#) sirver para crear encabezados. Se añade uno por cada nivel.

# Encabezado 1
## Encabezado 2
### Encabezado 3
#### Encabezado 4
##### Encabezado 5
###### Encabezado 6


## CITAS

Las citas se obtienen con el símbolo mayor que (>) al comienzo de un bloque de texto.

> Un país, una civilización se puede juzgar por la forma en que trata a sus animales.  —
Mahatma Gandhi


Si la cita se compone de varios párrafos, se debe añadir el símbolo > al comienzo de cada uno de ellos.

> Creo que los animales ven en el hombre un ser igual a ellos que ha perdido de forma extraordinariamente peligrosa el sano intelecto animal.

> Es decir, que ven en él al animal irracional, al animal que ríe, al animal que llora, al animal infeliz. — Friedrich Nietzsche

Inclusive se puede generar citas anidadas.

> Esto sería una cita como la que acabas de ver.
> 
> > Dentro de ella puedes anidar otra cita.
> 
> La cita principal llegaría hasta aquí. 



## LISTAS

### Listas desordenadas

Se utiliza asterisco (*), guiones (-), o símbolo de suma (+).

- Elemento de lista 1
- Elemento de lista 2
* Elemento de lista 3
* Elemento de lista 4
+ Elemento de lista 5
+ Elemento de lista 6


Para generar listas anidadas,simplemente se añaden **cuatro espacios en blanco antes del siguiente *. -, o +

- Elemento de lista 1
- Elemento de lista 2
    - Elemento de lista 3
    - Elemento de lista 4
        - Elemento de lista 5
        - Elemento de lista 6
        

### Listas ordenadas

Las listas ordenadas se crean usando la sintaxis tipo número (1.). También se pueden anidar o combinar.

1. Elemento de lista 1
2.  Elemento de lista 2
    - Elemento de lista 3
    - Elemento de lista 4
        1. Elemento de lista 5
        2. Elemento de lista 6
        
        
## CÓDIGOS DE BLOQUE

Para crear un bloque que contenga código, lo que se tiene que hacer es encerrar el párrafo entre dos líneas
formadas por tres virgulillas (~)

~~~
Creando códigos de bloque.
Puedes añadir tantas líneas y párrafos como quieras.  
~~~


## LÍNEAS HORIZONTALES

Se usan para separar secciones de una manera visual. Se crean cuando se incluyen tres de los siguiente elementos:  
asteriscos (*), guiones (-) o guiones bajos (_).

***
---
___



# ELEMENTOS DE LÍNEA

## ÉNFASIS

Se utilizan asteriscos o guiones bajos para enfatizar
Simplemente se envuelven las palabras o texto en estos símbolos para conseguir cursivas o negritas.

*cursiva*

_cursiva_

**negrita**

__negrita__



Se puede combinar la sintaxis, envolviendo la palabra en tres asteriscos o tres guiones bajos.

***cursiva y negrita***

___cursiva y negrita___


## LINKS O ENLACES


Se crean escribiendo la palabra o texto enlazada entre corchetes y el link en cuestión entre paréntesis.

[Colegio de Postgraduados](http://www.colpos.mx)

### Links o enlaces como referencia

El método anterior puede ser complicado cuando se escriben links largos o complejos. Para esto se pueden generar  
enlaces de referencia.

Esto quiere decir que en el texto se enlazarán palabras o códigos concretos que en otra sección del documento  
estarán definidos como URL.

[Colegio de Postgraduados][CP]

[CP]: http:www.colpos.mx


## CÓDIGO

Cuando se requiere añadir secciones donde mostrar código de otro lenguaje, atajos de teclado, o demás contenido que no debería ser tratado como tal.

### Código puro

Se envuelve el texto en dos comillas senciallas, que corresponde con al etiqueta HTML code
    
`Ejemplo de una línea de código`
    
    

    
### Texto preformateado  

La otra manera de añadir código en Markdown es comenzar el párrafo con cuatro espacios en blanco. Esto corresponde  
con la etiqueta HTML pre.


    Esto es una línea de código
    


## IMÁGENES

Se insertan imágenes de forma similar a insertar links solo que faltaría añadir el símbolo de exclamación al prinicipio y el enlace será la ubicación de la imagen.

# ELEMENTOS VARIOS

## LINKS AUTOMÁTICOS


Hay tipos de links automáticos. Estos son necesarios cuando se quiere mostrar una URL completa y no un enlace. Se geberan con los símbolos <>

<http://www.colpos.mx>


## OMITIR MARKDOWN

Para indicar que la escritura no se interprete como Markdown se escribe la barra invertida delante de calquiera de los elementos descritos anteriormente.

\  barra invertida

\`  acento invertido

\*  asterisco

\_  guión bajo

\{} llaves

\[] corchetes

\() paréntesis

\#  almohadilla

\+  símbolo de suma

\-  guión

\.  punto

\!  exclamación
