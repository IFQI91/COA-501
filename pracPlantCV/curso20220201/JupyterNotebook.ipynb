# JUPYTER NOTEBOOK (Jupyterlab)


Es un sistema para desarrollar programas de libre acceso y código abierto, con estándares abiertos y comunicación interactiva a lo largo de docenas de lenguajes de programación.

Jupyter notebook es una aplicación en la web de código abierto que permite crear y compartir documentos que contienen código vivo, ecuaciones, visualizaciones y texto narrativo.




Características:

- Edición de código en la web, con formato de texto.
- Ejecución del código desde el navegador.
- Presenta los resultados de los cálculos en HTML, LaTex, PNG, SVG, etc.
- Usa el el editor de texto Markdown.
- Facilita la inclusión de notación matemática en celdas markdown con LaTeX, y acceso nativo en MathJax.

Los documentos internamente están en archivos JSON (acrónimo de JavaScript Object Notation, 'notación de objeto de JavaScript') y se guardan con la extensión .ipynb. Como JSON es un archivo sin formato, se puede compartir.


Los cuadernos se pueden exportar a diversos formatos, incluyendo HTML (para posteo a un blog), reStructuredText, LaTeX, PDF, y presentación de láminas con el comando nbconvert.


## Privacidad

Jupyter notebook se ejecuta en su propia computadora. Si la dirección es http://localhost: o http://127.0.0.1: es que su computadora está funcionando como servidor.El programa no envía datos locales a ningún otro lugar.


## INICIO DEL SERVIDOR


Se inicia el programa en la **línea de comando** con la siguiente instrucción:


jupyter notebook



Entra uno automáticamente al "dashboard" el cual presenta los cuadernos disponibles en el directorio.


### Crear una libreta
Se crean desde el dashboard con el botón 'New Notebook'

### Abrir una libreta
Se puede abrir una libreta desde la línea de comando o con la opción File/Open

Nota: Se puede abrir más de un servidor al mismo tiempo si se quiere trabajar en libretas ubicadas en diferentes directorios. La primera libreta se abre en el servidor con puerto 8888, y las siguientes libretas se abren en puertos cercanos a éste.



## ESTRUCTURA DE UNA LIBRETA


La libreta consiste de una secuencia de celdas. Cada celda es un campo multitexto de entradas, y y se ejecutan con el comando shift-enter o presionando el botón "play", "cell" o "Run". Hay tres tipos de celdas:


### Celdas de código

Permiten editar y escribir código. El lenguaje de programación depende del "kernel" pero el default es (IPython).

IPython se refiere a python interactivo.

El resulado de correr el código se presenta en una celda que dice "output". La salida no se limita a texto. También puede ser: imágenes de matplotlib, figuras, tablas HTML, etc. Esto se conoce como Capacidad Rica de despliegue de información de IPython.

### Celdas Markdown

El proceso computacional se puede enriquecer usando texto en Markdown.

En las celdas de tipo Markdown, se puede incluir ecuaciones matemáticas de una manera fácil, usando notación LaTeX estándar: $...$ para ecuaciones mátemáticas en línea y $$...$$ para presentar las ecuaciones matemáticas. 

### Celdas crudas

Se puede escribir en esta celdas las salidas directamente, y no son evaluadas en la libreta.


Nota: Para interrumpir un cálculo que está tomando mucho tiempo se puede usar la opción Kernel/Interrupt. De forma similar se puede usar la opción Kernel/restart.

## NAVEGADORES COMPATIBLES

Se puede usar Jupyter notebook en las versiones recientes de los siguientes navegadores:

- Chrome
- Safari
- Firefox

También puede funciones en las versiones recientes de Opera y Edge, pero es preferible usar  los navegadores indicados anteriormente.
